```julia
function heatmapBiPlot(isSave=false, isBW=false)
    if isBW
        clist = [:white, :black]
        str1 = "BlackWhite"
    else
        clist = [:white, :yellow, :red, :purple, :blue]
        str1 = "Color"
    end

    n = 32
    dU = 1.25
    U = Bidiagonal(dU*ones(n), -1*ones(n-1), :U)
    A = inv(U)
    gr()
    fig = heatmap(1:size(A,1),
        1:size(A,2), A,
        c=cgrad(clist),
        xlabel="Columns", ylabel="Rows",
        title=L"Heat map of $U_{32}^{-1}(1.25, -1)$", dpi=150, yflip=true)
    display(fig)
    if isSave
        savefig(fig, "./figure/plotsfig/Bi32$str1$dU.png")
    end
end

heatmapBiPlot(true, true)

function heatmapTriPlot(isSave=false, isBW=false)
    if isBW
        clist = [:white, :black]
        str1 = "BlackWhite"
    else
        clist = [:white, :yellow, :red, :purple, :blue]
        str1 = "Color"
    end

    n = 32
    dT = 2.25
    T = Tridiagonal(-1*ones(n-1), 2.25*ones(n), -1*ones(n-1))
    A = inv(T)
    gr()
    fig = heatmap(1:size(A,1),
        1:size(A,2), A,
        c=cgrad(clist),
        xlabel="Columns", ylabel="Rows",
        title=L"Heat map of $T_{32}^{-1}(-1, 2.25, -1)$", dpi=150, yflip=true)
    display(fig)
    if isSave
        savefig(fig, "./figure/plotsfig/Tri32$str1$dT.png")
    end
end

heatmapTriPlot(true, true)
```
