```julia
U = generateTestTriangular(20, 1, Matrix, Float64, true)
n = 32


dU = 0.9
# for dU less than 1

U = Bidiagonal(dU*ones(n), -1*ones(n-1), :U)

gr()
Uinv = inv(U)
data = Uinv
# display(data)
fig1 = heatmap(1:size(data,2),
    1:size(data,1), data,
    c=cgrad([:white, :yellow, :red, :purple, :blue]),
    xlabel="x values", ylabel="y values",
    title="My title", dpi=150, yflip=true)
display(fig1)


dU = 1.1
# for dU great than 1

U = Bidiagonal(dU*ones(n), -1*ones(n-1), :U)

gr()
Uinv = inv(U)
data = Uinv
# display(data)
fig1 = heatmap(1:size(data,2),
    1:size(data,1), data,
    c=cgrad([:white, :yellow, :red, :purple, :blue]),
    xlabel="x values", ylabel="y values",
    title="My title", dpi=150, yflip=true)
display(fig1)

dU = 1.0
# for dU = 1

U = Bidiagonal(dU*ones(n), -1*ones(n-1), :U)

gr()
Uinv = inv(U)
data = Uinv
# display(data)
fig1 = heatmap(1:size(data,2),
    1:size(data,1), data,
    c=cgrad([:white, :yellow, :red, :purple, :blue]),
    xlabel="x values", ylabel="y values",
    title="My title", dpi=150, yflip=true)
display(fig1)






dT = 2.25
# for dT greater than 2
U = Tridiagonal(-1*ones(n-1), dT*ones(n), -1*ones(n-1))

gr()
Uinv = inv(U)
data = Uinv
# display(data)
fig1 = heatmap(1:size(data,2),
    1:size(data,1), data,
    c=cgrad([:white, :yellow, :red, :purple, :blue]),
    xlabel="x values", ylabel="y values",
    title="My title", dpi=150, yflip=true)
display(fig1)

dT = 1.5
# for dT less than 2
U = Tridiagonal(-1*ones(n-1), dT*ones(n), -1*ones(n-1))

gr()
Uinv = inv(U)
data = Uinv
# display(data)
fig1 = heatmap(1:size(data,2),
    1:size(data,1), data,
    c=cgrad([:white, :yellow, :red, :purple, :blue]),
    xlabel="x values", ylabel="y values",
    title="My title", dpi=150, yflip=true)

display(fig1)

dT = 2
# for dT = 2
U = Tridiagonal(-1*ones(n-1), dT*ones(n), -1*ones(n-1))

gr()
Uinv = inv(U)
data = Uinv
# display(data)
fig1 = heatmap(1:size(data,2),
    1:size(data,1), data,
    c=cgrad([:white, :yellow, :red, :purple, :blue]),
    xlabel="x values", ylabel="y values",
    title="My title", dpi=150, yflip=true)

display(fig1)


```
